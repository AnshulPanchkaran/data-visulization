# heatmap

```python
flights = sns.load_dataset('flights')
```

```python
flights.head()
```

```python
flights.pivot_table(index='year',columns='month',values='passengers',aggfunc='sum')
```

```python
x=flights.pivot_table(index='year',columns='month',values='passengers',aggfunc='sum')
```

```python
sns.heatmap(x)
```

```python
sns.heatmap(x,cbar= False,linewidth=0.5,annot= True,fmt='d')
```

```python
sns.heatmap(x,cbar= False,linewidth=0.5,annot= True,fmt='d',cmap='winter')
```

#clustermap

```python
sns.clustermap(x)
```

```python
sns.clustermap(x,metric='correlation')
```

```python
sns.clustermap(x,z_score=0,annot= True,row_cluster=False,metric='correlation')
```

#distplot are used to do univariate analysis on continous numerical data 
distplot combine 3 things in one graph
1.histogram
2.kernel density estimation(pdf)
3.rugplot

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
titanic['age'].fillna(titanic['age'].mean(),inplace= True)
```

```python
sns.distplot(titanic['age'])
```

```python
sns.distplot(titanic['age'],kde=False)
```

```python
sns.distplot(titanic['age'],hist=False)
```

```python
sns.distplot(titanic['age'],hist=False,kde=False,rug=True)
```

```python
sns.distplot(titanic[titanic['survived']==1]['age'])
sns.distplot(titanic[titanic['survived']==0]['age'])
```
